```python

# 1. Load your Cartesian NetCDF (created in the previous steps)
# If you haven't made it yet, use your existing 'ds' from the RANS output
ds = xr.open_dataset('flow_gaussian_cartesian.nc')

# Extract the components
# Assuming dimensions are (x, y, z)
U = ds.U.values * 8
V = ds.V.values * 8
W = ds.W.values * 8
x_vec = ds.x.values
y_vec = ds.y.values
z_vec = ds.z.values

# 2. Setup the figure
fig, ax = plt.subplots(figsize=(10, 8))

# Initial slice index (middle of the height)
initial_idx = 8 # len(z_vec) // 2

# 3. Create the initial plot
# We transpose U[initial_idx] if needed to match (y, x) for pcolormesh
X, Y = np.meshgrid(x_vec, y_vec)
# We plot U[x, y, initial_idx]
mesh = ax.pcolormesh(X, Y, W[:, :, initial_idx].T, shading='auto', cmap='viridis')

# Add formatting
plt.colorbar(mesh, ax=ax, label='Velocity $U$ [m/s]')
ax.set_title(f'Wind Velocity at Height: {z_vec[initial_idx]:.2f}m')
ax.set_xlabel('X [m]')
ax.set_ylabel('Y [m]')
ax.set_aspect('equal')
plt.tight_layout()
plt.show()
```
